```python
import cv2

def main():
    # Initialize the webcam
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Error: Could not open webcam.")
        return

    # Load the Haar Cascade classifier
    face_cascade_path = r"C:/Users/91708/Desktop/harra/haarcascade_frontalface_default.xml"
    faceCascade = cv2.CascadeClassifier(face_cascade_path)
    if faceCascade.empty():
        print("Error: Could not load Haar Cascade.")
        return

    while True:
        # Capture frame-by-frame
        success, img = cap.read()
        if not success:
            print("Error: Could not read frame.")
            break

        # Detect faces in the image
        faces = faceCascade.detectMultiScale(img, scaleFactor=1.2, minNeighbors=4)

        # Blur detected faces
        for (x, y, w, h) in faces:
            ROI = img[y:y+h, x:x+w]
            blur = cv2.GaussianBlur(ROI, (91, 91), 0)
            img[y:y+h, x:x+w] = blur

        # Display message if no faces are detected
        if len(faces) == 0:
            cv2.putText(img, 'No Face Found!', (20, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255))

        # Display the resulting frame
        cv2.imshow('Face Blur', img)

        # Break the loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the webcam and close all OpenCV windows
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
